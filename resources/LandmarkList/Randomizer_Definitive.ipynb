```python
import pandas as pd
import random

df = pd.read_csv('Landmarks_English_CSV.csv', encoding='ISO-8859-1')
print('DF:')
print(df)
print('----------------------\n')

n_elements_id = 30
n_elements_country = 7
n_elements_continent = 2

def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')
    
# print_full(pd.concat([df, mask_global], axis=1).iloc[:,1:])

def country_remover(df,id_day):
    # This function creates a mask that returns the countries that are the same country as id_day (the id)
    # The id_day is suposed to be a list
    mask_country = df.loc[:,'Country'] != df.loc[:,'Country'] # Inicializando el bool
    for i_id_day in id_day:
        print('Country remover function')
        print('i_id_day: ', i_id_day, '. Country: ', df.loc[i_id_day,'Country'])
        mask_country_dummy = df.loc[:,'Country'] == df.loc[:,'Country'][df.loc[:,'id'] == i_id_day].iloc[0]
        mask_country = mask_country | mask_country_dummy
    return mask_country

def continent_remover(df,id_day):
    # This function creates a mask that returns the CONTINENT that are the same country as id_day (the id)
    # The id_day is suposed to be a list
    mask_country = df.loc[:,'Continent'] != df.loc[:,'Continent'] # Inicializando el bool
    for i_id_day in id_day:
        print('Continent remover function')
        print('i_id_day: ', i_id_day, '. Continent: ', df.loc[i_id_day,'Continent'])
        mask_country_dummy = df.loc[:,'Continent'] == df.loc[:,'Continent'][df.loc[:,'id'] == i_id_day].iloc[0]
        mask_country = mask_country | mask_country_dummy
    return mask_country

def id_remover(df,id_day):
    # This function creates a mask that returns the ID that are the same country as id_day (the id)
    # The id_day is suposed to be a list
    mask_country = df.loc[:,'id'] != df.loc[:,'id'] # Inicializando el bool
    for i_id_day in id_day:
        print('id remover function')
        print('i_id_day: ', i_id_day)
        mask_country_dummy = df.loc[:,'id'] == i_id_day
        mask_country = mask_country | mask_country_dummy
    return mask_country

def global_mask(df, landmark_list,n_elements_id,n_elements_country,n_elements_continent):  
    selection_list_reversed = list(reversed(landmark_list))

    mask_country = country_remover(df,selection_list_reversed[0:n_elements_country])
    mask_continent = continent_remover(df,selection_list_reversed[0:n_elements_continent])
    mask_id = id_remover(df,selection_list_reversed[0:n_elements_id])
    
    mask_global = mask_country | mask_continent | mask_id
    return mask_global
```

```python
def randomizer(df,landmark_list,limit):
    flag_dif = 0 #Flag for the difficulty
    # for i in range(limit):
    while len(landmark_list) < limit:
        print('Before Randomizer: landmark id list:', landmark_list)
        
        # Masking
        mask_global = global_mask(df, landmark_list,n_elements_id,n_elements_country,n_elements_continent)

        # Creting next df
        next_df = df[~mask_global]

        # Obtaining next Landmark
        new_landmark_id = random.choice(next_df['id'].tolist())
        print('New landmark id: ', new_landmark_id)
        
        # Appeinding the new landmark to the list
        landmark_list.append(new_landmark_id)
        flag_dif = flag_dif + 1
        print('After Randomizer: landmark id list:', landmark_list)
        print('Length landmark id list:', len(landmark_list))
        if flag_dif == 7:
            flag_dif = 0 # Resseting flag
            print('DIFFICULTY ASSESSMENT')
            average_param = (df.loc[landmark_list[-7:] ,'Dif']).mean()
            if average_param>=1: # way too difficult
                print('The landmarks were too difficult:')
                print('average_param:', average_param)
                print('Landmarks:', df.loc[landmark_list[-7:],['Name','Dif']])
                print('Resetting...')
                landmark_list = landmark_list[:-7]
            else:
                print('The landmarks were not too difficult. Mean:', average_param)
                print('Landmarks:', df.loc[landmark_list[-7:],['Name','Dif']])
                pass
    return landmark_list
```
